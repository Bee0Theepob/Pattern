```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def gaussian(x, mean, sigma_2):
    return (1 / (math.sqrt(2 * math.pi * sigma_2))) * np.exp(-(x - mean)**2 / (2 * sigma_2))
```

```python
X=np.arange(-5,10,0.001)
Y=[]
Y2=[]
for x in X:
    Y.append(gaussian(x,4,2)*0.5)
Y=np.array(Y)
plt.plot(X,Y,'-',color="red")
for x in X:
    Y2.append(gaussian(x,0,2)*0.5)
Y2=np.array(Y2)    

plt.plot(X,Y2,'-',color="blue")
plt.grid()
plt.show()
print("T2) dicision boundary is 2")
```

```python
X=np.arange(-5,10,0.001)
Y=[]
Y2=[]
for x in X:
    Y.append(gaussian(x,4,2)*0.75)
Y=np.array(Y)
plt.plot(X,Y,'-',color="red")
for x in X:
    Y2.append(gaussian(x,0,2)*0.25)
Y2=np.array(Y2)  

print("T3")
plt.plot(X,Y2,'-',color="blue")
plt.grid()
plt.show()
```
